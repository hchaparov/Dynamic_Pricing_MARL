# Baselines

```python
import matplotlib.pyplot as plt

# Data
agents = ['Negamax', 'Random']
win_rates = [0.98, 0.02]

# Create plot
fig, ax1 = plt.subplots(figsize=(7, 6))  # Adjust figsize if needed

# Plot win rates
bars1 = ax1.bar(agents, win_rates, color=['blue', 'orange'])
ax1.set_title('Win Rates of Agents')
ax1.set_xlabel('Baseline opponents', fontweight='bold')
ax1.set_ylabel('Win rate', fontweight='bold')
ax1.set_ylim(0, 1)  # Win rates are between 0 and 1
ax1.yaxis.grid(True)

# Add win rate percentages above bars
for bar in bars1:
    yval = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, yval + 0.01, f'{yval:.0%}', ha='center', va='bottom')

# Show plot
plt.tight_layout()
plt.show()
```

# 2A_Connect4_PPO

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['PPO vs Random agent', 'PPO vs Negamax agent', 'PPO vs PPO agent']
agent1_win_percentages = [0.84, 0.06, 0.57]
agent2_win_percentages = [0.16, 0.94, 0.43]

x = np.arange(len(categories))  # the label locations
width = 0.35  # the width of the bars

# Plotting win percentages
fig, ax = plt.subplots(figsize=(10, 6))

rects1 = ax.bar(x - width/2, agent1_win_percentages, width, label='200 000-timestep PPO agent Win %', color='blue')
rects2 = ax.bar(x + width/2, agent2_win_percentages, width, label='Random/Negmax/PPO agent Win %', color='orange')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Opponents', fontweight='bold')
ax.set_ylabel('Win rate', fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories)
ax.set_ylim(0, 1)
ax.legend()

# Function to add labels on top of bars
def autolabel(rects, ax):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.0%}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1, ax)
autolabel(rects2, ax)

fig.tight_layout()

plt.show()
```

# MCTS

```python
import matplotlib.pyplot as plt

# Data
strategies = ['MCTS vs Random', 'MCTS vs Negamax']
win_rates = [100, 98]
draw_rates = [0, 2]
loss_rates = [0, 0]  # Assuming 0% loss rates since it's not mentioned

# Bar width
bar_width = 0.25

# Bar positions
r1 = range(len(strategies))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

# Create bar plot
bars1 = plt.bar(r1, win_rates, color='b', width=bar_width, edgecolor='grey', label='Win Rate of\n MCTS agent')
bars2 = plt.bar(r2, draw_rates, color='g', width=bar_width, edgecolor='grey', label='Draw Rate of\n MCTS agent')
bars3 = plt.bar(r3, loss_rates, color='orange', width=bar_width, edgecolor='grey', label='Loss Rate of\n MCTS agent')

# Add labels
plt.xlabel('Opponents', fontweight='bold')
plt.ylabel('Percentage', fontweight='bold')
plt.xticks([r + bar_width for r in range(len(strategies))], strategies)
plt.title('')

# Add percentage text on bars
for bar in bars1:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}%', ha='center', va='bottom')

for bar in bars2:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}%', ha='center', va='bottom')

for bar in bars3:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, f'{yval}%', ha='center', va='bottom')

# Add legend
plt.legend()

# Show plot
plt.show()
```
